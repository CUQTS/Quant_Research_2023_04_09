```python
import os
import pandas as pd
import numpy as np
from scipy.stats import t
```

```python
Backtest_Result_Path = 'C:\\Users\\anson\\OneDrive\\桌面\\CUQTS\\2023_04_03\\Backtest Result\\result_df.xlsx'
result_df = pd.read_excel(Backtest_Result_Path, index_col = 0)
```

```python
for index in result_df.index : 
    print(result_df.loc[index])
```

```python
# separate the performance by timeframe

annualized_ret_1M = []
sharpe_1M = []
index_set_1M = []
annualized_ret_3M = []
sharpe_3M = []
index_set_3M = []

for index in result_df.index:
    if index[-2:] == "1M" :
        annualized_ret_1M.append(result_df.loc[index]['Annualized Return'])
        sharpe_1M.append(result_df.loc[index]['Sharpe Ratio'])
        index_set_1M.append(index)
    else:
        annualized_ret_3M.append(result_df.loc[index]['Annualized Return'])
        sharpe_3M.append(result_df.loc[index]['Sharpe Ratio'])
        index_set_3M.append(index)
```

```python
annualized_ret_1M_df = pd.DataFrame(annualized_ret_1M, index = index_set_1M)
sharpe_1M_df =  pd.DataFrame(sharpe_1M, index = index_set_1M)

annualized_ret_3M_df = pd.DataFrame(annualized_ret_3M, index = index_set_3M)
sharpe_3M_df =  pd.DataFrame(sharpe_3M, index = index_set_3M)
```

```python
# statistics

print("1M Annualized Ret Mean : ", annualized_ret_1M_df.mean())
print("1M Annualized Ret STD : ", annualized_ret_1M_df.std())
print("1M Sharpe Mean : ", sharpe_1M_df.mean())
print("1M Sharpe STD : ", sharpe_1M_df.std())

print("3M Annualized Ret Mean : ", annualized_ret_3M_df.mean())
print("3M Annualized Ret STD : ", annualized_ret_3M_df.std())
print("3M Sharpe Mean : ", sharpe_3M_df.mean())
print("3M Sharpe STD : ", sharpe_3M_df.std())
```

```python
# hypothesis testing on annualized return and sharpe ratio

print('1M Annualized Ret p-value :', t.cdf(annualized_ret_1M_df.mean()/(annualized_ret_1M_df.std()/np.sqrt(len(annualized_ret_1M_df))), len(annualized_ret_1M_df)-1))
print('1M Sharpe p-value :', t.cdf(sharpe_1M_df.mean()/(sharpe_1M_df.std()/np.sqrt(len(sharpe_1M_df))), len(sharpe_1M_df)-1))

print('3M Annualized Ret p-value :', t.cdf(annualized_ret_3M_df.mean()/(annualized_ret_3M_df.std()/np.sqrt(len(annualized_ret_3M_df))), len(annualized_ret_3M_df)-1))
print('3M Sharpe p-value :', t.cdf(sharpe_3M_df.mean()/(sharpe_3M_df.std()/np.sqrt(len(sharpe_3M_df))), len(sharpe_3M_df)-1))
```

```python
# Top performance

print(sharpe_1M_df.sort_values(by = [0], ascending = False).head(3))
print(sharpe_3M_df.sort_values(by = [0], ascending = False).head(3))
```
